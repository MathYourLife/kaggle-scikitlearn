```python
train = np.genfromtxt('../data/train.csv', delimiter=',')
trainLabels = np.genfromtxt('../data/trainLabels.csv', delimiter=',', dtype=int)
test = np.genfromtxt('../data/test.csv', delimiter=',')
```

```python
train2 = None
for row in train:
    squares = np.array([])
    for val in row:
        squares = np.append(squares, val**2)
    row = np.append(row, squares)
    if train2 is None:
        train2 = row
    else:
        train2 = np.vstack((train2, row))
```

```python
from sklearn.svm import LinearSVC
clf = LinearSVC()
clf = clf.fit(train2, trainLabels)
validate = clf.predict(train2)
print sum(abs(validate - trainLabels))
testLabels = clf.predict(test)
```
